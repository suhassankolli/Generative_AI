```python
import pandas as pd
import numpy as np

def compute_stats(df, cols=[]):  # mutable default
    stats = {}
    for c in cols:
        stats[c] = {"mean": df[c].mean(), "std": df[c].std()}
    return stats

def slow_loop(df):
    out = []
    for i,row in df.iterrows():   # slow
        out.append(row["x"] * 2)
    return out

def risky(x):
    try:
        return eval(x)   # unsafe
    except Exception:
        return None  # hides errors

def very_long_function(df):
    # Intentionally too long and nested loops to trigger maintainability/perf findings
    total = 0
    for i in range(len(df)):
        for j in range(len(df)):
            total += df.iloc[i,0] * df.iloc[j,0]
    return total
```
